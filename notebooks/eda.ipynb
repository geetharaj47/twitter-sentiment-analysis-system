```python
import pandas as pd
```

```python
columns = [
    "target",
    "id",
    "date",
    "flag",
    "user",
    "text"
]

df = pd.read_csv(
    "../data/raw/training.1600000.processed.noemoticon.csv",
    encoding="latin-1",
    names=columns
)

df.head()
```

```python
df["target"].value_counts()
```

```python
df["target"] = df["target"].replace(4,1)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["text"].head(10)
```

```python
df["tweet_length"] = df["text"].apply(len)
```

```python
import matplotlib.pyplot as plt

plt.hist(df["tweet_length"], bins=50)
plt.title("Tweet Length Distribution")
plt.xlabel("Length")
plt.ylabel("Frequency")
plt.show()
```

```python
df["target"].value_counts().plot(kind="bar")
plt.title("Sentiment Distribution")
plt.show()
```

```python
df_sample = df.sample(100000, random_state=42)
df_sample.to_csv("../data/processed/sample_tweets.csv", index=False)
```
